# 3.2 Observe Distributions

```python
numerical = ['age','education','hours_per_week']
categorical = ['workclass','marital_status','occupation','relationship',
               'race','sex','capital_change','native_country']
```

```python
sns.set(rc={'figure.figsize':(20,10)})

for feature in numerical:
    sns.countplot(data=train, x=feature, color='lightcoral')
    plt.show()
```

```python
# 'age' shows a similiar distribution to the population
# no of data counts decrease with education level
# most data has 40 hours_per_week
```

```python
sns.set(rc={'figure.figsize':(20,10)})

for feature in categorical:
    sns.countplot(data=train, x=feature, color='lightcoral')
    plt.show()
```

```python
# 'native_country' has too many categories
# rotate labels for clearer view

sns.set(rc={'figure.figsize':(20,10)})

ax = sns.countplot(data=train, x='native_country', color='lightcoral')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

```python
# most data's 'workclass' is 'Private'
# most data's 'race' is 'White'
# most data has no 'capital_change'
# nearly all data's native country is 'United State'
```

# 3.3 Observe Distributions with 'target'

```python
# relationship between 'age' and 'target'

plt.figure(figsize=(20,10))
sns.countplot(data=train, x='age', hue='target', palette='Reds')
```

```python
# most data with over 50k annual income has an age between 30 to 60
```

```python
# relationship between 'education' and 'target'

plt.figure(figsize=(20,10))
sns.countplot(data=train, x='education', hue='target', palette='Reds')
```

```python
# the observation is interesting
# considering the 4 most common education levels: from 0 to 4
# each category has similiar numbers of data with over 50k annual income
# that means with PhD graduates excluded
# high income individuals distribute even across different education levels
```

```python
# relationship between 'hours_per_week' and 'target'

plt.figure(figsize=(20,10))
sns.countplot(data=train, x='hours_per_week', hue='target', palette='Reds')
```

```python
# high income is associated with long working hours
```

```python
# relationship between 'marital_status' and 'target'

plt.figure(figsize=(20,10))
sns.countplot(data=train, x='marital_status', hue='target', palette='Reds')
```

```python
# married people has higher percentage of over 50k annual income
# we cannot be certain about whether it is correlation or causation relationship
# in this project we only consider the correlation of features and target
# further study is required for causal inference
```

```python
# relationship between 'race' and 'target'

plt.figure(figsize=(20,10))
sns.countplot(data=train, x='race', hue='target', palette='Reds')
```

```python
# with sufficient data
# black has larger portion of data with below 50k annual income
# points to note: Asian-Pac-Islander group of data has highest percentage of high income data
```

```python
# relationship between 'sex' and 'target'

plt.figure(figsize=(20,10))
sns.countplot(data=train, x='sex', hue='target', palette='Reds')
```

```python
# significantly less female with over 50k annual income
```

```python
# relationship between 'capital_change' and 'target'

plt.figure(figsize=(20,10))
sns.countplot(data=train, x='capital_change', hue='target', palette='Reds')
```

```python
# 3.5 encoding categorical features

# have to put it in the modelling part
```

```python
numerical_features = ['age','education','hours_per_week']

numerical_transformer = StandardScaler()
# if more than one step
# use Pipeline(steps=[])
```

```python
categorical_features = ['workclass','marital_status','occupation','relationship',
               'race','sex','capital_change','native_country']

categorical_transformer = OneHotEncoder(drop='first')
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num', numerical_transformer, numerical_features),
    ('cat', categorical_transformer, categorical_features)
])
```

```python
x = preprocessor.fit_transform(train)
```
